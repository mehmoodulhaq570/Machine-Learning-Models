## Categorization of Spam and Not Spam Emails 

```python
import pandas as pd
df = pd.read_csv("C://Users//mehmo//Downloads//Machine Learnig Course//14 Naive Bayes Classifier P2//spam.csv")
df.head()
```

```python
df.groupby("Category").describe()
```

```python
df['spam'] = df['Category'].apply(lambda x:1 if x=='spam' else 0)
df.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df.Message, df.spam, test_size=0.20)
```

```python
X_train
```

```python
X_train.values
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X_train_count = vectorizer.fit_transform(X_train.values)
X_train_count.toarray()[:3]
```

CountVectorizer is a popular feature extraction technique in sklearn used for converting text documents into numerical feature vectors. It counts the occurrences of each word in the document and represents the text as a matrix of token counts.

```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
emails = [
    'Hey brother how are you doing today? I hope that you are aware we have cricket match today',
    'You have got a prize of 500 $ call our customer service to avail the reward',
    'Where were you from the previous day. We have to celebrate our success'
]
email_count = vectorizer.transform(emails)
model.predict(email_count)
```

```python
X_test_count = vectorizer.transform(X_test)
model.score(X_test_count, y_test)
```

The concept of pipelines in scikit-learn is to combine multiple steps into a single model. Pipelines are useful when you have a sequence of transformations that need to be applied to your data before feeding it into a machine learning model.

```python
# As you can see we have to perform the tranformation on data every time we are going to feed it into our model. 
# So to save ourself from this we are going to use a feature of sklearn called as pipeline

from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```

```python
clf.predict(emails)
```
