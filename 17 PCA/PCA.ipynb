```python
import numpy as np # type: ignore
np.unique(dataset.target) # classifying the data into 10 classes
```

```python
df = pd.DataFrame(dataset.data, columns = dataset.feature_names) # 1797 images are distributed among 10 classes
df.head()
```

```python
df.describe() #
```

```python
X = df
y = dataset.target
```

```python
from sklearn.preprocessing import StandardScaler # type: ignore
scalar = StandardScaler()
X_scaled = scalar.fit_transform(X) 
X_scaled
```

```python
from sklearn.model_selection import train_test_split # type: ignore
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size = 0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression # type: ignore
model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
X.shape
```

```python
from sklearn.decomposition import PCA # type: ignore
pca = PCA(0.95) # retain 95% of useful feature and then create new dimentions 
X_pca = pca.fit_transform(X)
X_pca.shape

# originally the shape of X is 64, after PCA it is reduced to 29
```

```python
pca.explained_variance_ratio_
# It show that in my 29 new dimentions, how much variance is explained by each dimentions
```

```python
pca.n_components_
```

```python
X_train_pca, X_test_pca, y_train, y_test  = train_test_split(X_pca, y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression(max_iter=1000)
model.fit(X_train_pca, y_train)
```

```python
model.score(X_test_pca, y_test)
```
