```python
dataset = pd.read_csv("C://Users//mehmo//Downloads//Machine Learnig Course//16 L1 and L2 regularization//Melbourne_housing_FULL.csv")
dataset.head()
```

```python
dataset = dataset.drop(['Address','Distance','Lattitude','Longtitude','Postcode','YearBuilt'],axis=1)
dataset
```

```python
dataset.shape
```

```python
# This will find the missing values in the dataset and sum the number of missing values in each column
dataset.isna().sum()
```

```python
# This code generate the list and asign the 0 value to the missing points in column and then return back the columns to the original dataset
cols_to_fill_zero = ['Propertycount', 'Bathroom', 'Bedroom2', 'Car']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
dataset.isna().sum()
```

```python
# This will drop those rows where values are missing
dataset.dropna(inplace=True)
dataset.isna().sum()
```

```python
# We have to drop first colums to prevent form the dummy variable trap
dataset = pd.get_dummies(dataset, drop_first=True)
dataset.head()
```

```python
X = dataset.drop('Price', axis=1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split # type: ignore
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size=0.3, random_state=2)
```

```python
from sklearn.linear_model import LinearRegression # type: ignore
reg = LinearRegression().fit(train_X, train_y)
```

```python
reg.score(test_X, test_y)
```

```python
# For traning it is giving the good results but for testing it is givig the low results. It means that the model is overfitting. 
# So we have to use the regularization technique to prevent the model from overfitting.
reg.score(train_X, train_y)
```

```python
# We are going to use L1 regularization also called as Lesso regularization
from sklearn import linear_model # type: ignore

lesso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1) # alpha as shrinking factor to prevent the overfitting of the model

lesso_reg.fit(train_X, train_y)
```

```python
# Here we can see that the model accuracy has been incresed from -0.59 to 0.67 by avoiding the overfitting using lesso regularization
lesso_reg.score(test_X, test_y)
```

```python
from sklearn import linear_model # type: ignore

ridge_reg = linear_model.Ridge(alpha=50, max_iter=100, tol=0.1) # alpha as shrinking factor to prevent the overfitting of the model 
#The tol parameter stands for tolerance. It is the stopping criterion.

ridge_reg.fit(train_X, train_y)
```

```python
ridge_reg.score(test_X, test_y)
```
