```python
# This model is used to predict the handwritten digit recognition using logistic regression multiclassification
# matplot inline method is used to plot the graph below the code cell without any need for further plotting configuration

%matplotlib inline
import matplotlib.pyplot as plt

# sklearn datasets have some predefined datasets that can be used to learn machine learning

from sklearn.datasets import load_digits
```

```python
digits = load_digits()
```

```python
dir(digits)
```

```python
# It is an array depiction of 8x8 D picture
digits.data[0]
```

```python
plt.gray()
for i in range (5):
    plt.matshow(digits.images[i])
```

```python
# The targeted sequence is the same as in the above picture

digits.target[0:5]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=5000)
```

```python
model.fit(X_train, y_train)
```

```python
# Form X_test data the model with predict the targeted value and then match those targeted values with the actual y_test values
model.score(X_test, y_test)
```

```python
plt.matshow(digits.images[67])
```

```python
# The targeted value for the above image is below.
digits.target[67]
```

```python
# As you can see the predicted value for the image is correct as the targeted value, because our accuracy is high
model.predict(digits.data[67].reshape(1,-1))
```

```python
model.predict(digits.data[0:5])
```

```python
# To get to know the feeling of your model accuracy here is what we used called a confusion matrix

y_predicted = model.predict(X_test) # This consists of our predicted value for X_test data 
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_predicted) # This consist of the our actual value and the predicted values and their matrix is being displayed.
cm
```

```python
import seaborn as sn
plt.figure(figsize=(10,7))
sn.heatmap(cm, annot = True) 
plt.xlabel("Pedicted Value")
plt.ylabel("True Value")

# The number in each matrix show the number of time wrong value predicted by model. Zero indicted that no errors. 
```
