```python
import pandas as pd
df = pd.read_csv("C://Users//mehmo//Downloads//Machine Learnig Course//13 Naive Bayes Classifier P1//titanic.csv")
df.head()
```

```python
# We are going to remove those features on which the survival of the passager does not depend
df.drop(['PassengerId', 'Name', 'SibSp','Ticket', 'Cabin','Embarked' ], axis='columns', inplace=True)
# Inplce=ture means it will not create a new dataframe, alternatively it will modiy the current dataframe
df.head()
```

```python
target = df.Survived
inputs = df.drop(['Survived'], axis='columns')
```

```python
dummies = pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
# Converting the boolean values to the numerical values
dummies = dummies.astype(int)
dummies.head(3)
```

```python
inputs = pd.concat([inputs, dummies], axis='columns')
inputs.head(3)
```

```python
inputs = inputs.drop(['Sex'],axis='columns')
inputs.head(3)                    
```

```python
inputs.columns[inputs.isna().any()]
# isna().any() is a pandas function that checks for missing values (NaN) in each column of the df inputs
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())
inputs.head(6)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs, target, test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
model.predict(X_test[:10])
```
